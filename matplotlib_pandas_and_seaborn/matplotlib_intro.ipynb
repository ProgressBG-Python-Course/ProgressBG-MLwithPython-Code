```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,16)
y = x**2

# lets plot:
plt.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, sharey=True)
fig.set_facecolor("#CCCCCC")

h, pl2 = axes

# data
x = np.arange(1,31)
y1 = 2*x
y2 = x/2

# customize: 
pl1.set_title("y = 2*x")
pl1.set_xlabel("X", fontsize=16, color="#FF0000" )
pl1.set_ylabel("Y", fontsize=16, color="#00FF00", rotation=0 )
pl1.xaxis.set_label_coords(1,0)
pl1.xaxis.set_label_coords(0,0)
pl2.set_title("y = 2*x") 


# plot
pl1.plot(x, y1, color='blue', linestyle=":")
pl2.plot(x, y2, color='green')
plt.tight_layout()
```
