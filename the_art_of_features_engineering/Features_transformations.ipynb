## Log Function

$\log_{b} (x) = y$, if $b^y = x$

```python
x = np.arange(1,100)

ylog10 = np.log10(x)
ylog2 = np.log2(x)
yln = np.log(x)
# print(f'x: {x}')
# print(f'y: {y}')
```

```python
plt.figure(dpi=200)
plt.plot(x,ylog2,color="red",label="y=log2(x)")
plt.plot(x,ylog10, color="green",label="y=log10(x)")
plt.plot(x,yln,color="blue",label="y=logE(x)")

plt.title('Logarithm Functions')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
